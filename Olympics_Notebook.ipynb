# create function for find countries glories.

```python
def countries_info(country, series='Edition'):
    the_country = olympics_data_set[olympics_data_set['NOC'] == country]
    if len(the_country) != 0:
        if series not in ('Edition', 'City', 'Medal', 'Sport', 'Athlete', 'Gender'):
            return f'{series} does not exist in olympics series!'
        else:
            return the_country.groupby(series).size()
    else:
        return f'{country} country does not exist in olympics history!'
```

```python
countries_info(country='USA', series='Medal')
```

# final challenge

### in every olympics, wich USA Athlete has won the most total number of medals? include discipline.

### my way:

```python
def first_champion(country, start_year=1896, end_year=2008, head=1):
    final_data_frame = pd.DataFrame()
    
    for year in range(start_year, end_year + 1):
        c = olympics_data_set[(olympics_data_set['Edition'] == year) & (olympics_data_set['NOC'] == country)]
        first_champ = c.groupby(['Edition', 'Athlete', 'Discipline', 'Medal']).size().unstack('Medal', fill_value=0)

        if 'Gold' in first_champ:
            g = first_champ['Gold']
        else: g = 0; first_champ['Gold'] = 0

        if 'Silver' in first_champ:
            s = first_champ['Silver']
        else: s = 0; first_champ['Silver'] = 0

        if 'Bronze' in first_champ:
            b = first_champ['Bronze']
        else: b = 0; first_champ['Bronze'] = 0

        first_champ['Total'] = g + s + b
        
        first_champ = first_champ.sort_values(['Gold', 'Silver', 'Bronze'], ascending=False)[['Gold', 'Silver', 'Bronze', 'Total']]
        
        first = first_champ.head(head)
        
        final_data_frame = final_data_frame.append(first)
    
    return final_data_frame
```

```python
first_champion('IRI', head=2)
```

### another way:

```python
def champion(country):
    gy = olympics_data_set[olympics_data_set['NOC'] == country]
    gy = gy.groupby(['Edition', 'Athlete', 'Medal']).size().unstack('Medal', fill_value=0)

    gy['Total'] = gy['Gold'] + gy['Silver'] + gy['Bronze']

    gy.reset_index(inplace=True)

    tu = [group.sort_values('Total', ascending=False)[:1] for year, group in gy.groupby('Edition')]
    top = pd.DataFrame()

    for data in tu:
        top = top.append(data)
    
    return top
```

```python
champion('IRI')
```
